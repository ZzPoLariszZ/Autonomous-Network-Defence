```python
import inspect
import random
from pprint import pprint
from CybORG import CybORG

path = str(inspect.getfile(CybORG))
path = path[:-10] + '/Shared/Scenarios/Scenario1b.yaml'

env = CybORG(path, 'sim')

results = env.reset(agent='Red')
red_obs = results.observation
# pprint(red_obs)
pprint(list(results.action_space['action']))
# pprint(list(results.action_space['port']))
```

```python
pprint(red_obs)
print(76*'-')
blue_obs = env.get_observation('Blue')
pprint(blue_obs)
```

```python
from CybORG.Agents import B_lineAgent
from CybORG.Agents import CovertAgent

action_space = results.action_space
# b_line_agent = B_lineAgent()
b_line_agent = CovertAgent()

def step_red(red_obs):
    action = b_line_agent.get_action(red_obs, action_space)
    print('B-line Action:', action)
    print('B-line Action Num:', b_line_agent.action)
    print('B-line Target IP Address:', b_line_agent.target_ip_address)
    print('B-line Last Subnet:', b_line_agent.last_subnet)
    print('B-line Last IP Address:', b_line_agent.last_ip_address)
    # pprint(b_line_agent.action_history)
    print(76*'-')

    results = env.step(action=action, agent='Red')
    red_obs = results.observation
    pprint(red_obs)
    pprint(env.get_rewards())
    
    return results

results = step_red(red_obs)
red_obs = results.observation
print(76*'-')
blue_obs = env.get_observation('Blue')
pprint(blue_obs)
```
